```python
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers

# Generate synthetic time series data
def generate_synthetic_data(num_samples, num_steps):
    # Sinusoidal data with some noise
    time = np.arange(0, num_steps * num_samples)
    data = np.sin(0.1 * time) + np.sin(0.05 * time) + np.random.normal(scale=0.5, size=num_steps * num_samples)
    data = data.reshape((num_samples, num_steps))
    return data

# Create sequences from the generated data
def create_sequences(data, num_steps):
    X = []
    y = []
    for i in range(len(data) - num_steps):
        X.append(data[i:i+num_steps])
        y.append(data[i+num_steps])
    return np.array(X), np.array(y)

# Parameters
num_samples = 10000
num_steps = 50
batch_size = 32

# Generate and prepare the data
data = generate_synthetic_data(num_samples, num_steps + 1)
X, y = create_sequences(data, num_steps)

# Split the data into training and validation sets
train_size = int(0.8 * len(X))
X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:], y[train_size:]

# Reshape the data for the transformer model
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

# Define the transformer model
def transformer_encoder(inputs):
    # Normalization and attention
    x = layers.LayerNormalization(epsilon=1e-6)(inputs)
    attention_output = layers.MultiHeadAttention(num_heads=2, key_dim=2)(x, x)
    x = layers.Add()([x, attention_output])
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    # Feed-forward network
    x = layers.Conv1D(filters=64, kernel_size=1, activation='relu')(x)
    x = layers.GlobalMaxPooling1D()(x)
    return x

# Build the model
inputs = keras.Input(shape=(num_steps, 1))
x = transformer_encoder(inputs)
outputs = layers.Dense(1)(x)
model = keras.Model(inputs=inputs, outputs=outputs)

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, y_train, batch_size=batch_size, epochs=10, validation_data=(X_val, y_val))

# Forecasting function
def forecast(model, data, num_steps):
    predictions = model.predict(data)
    return predictions.flatten()

# Use the last sequence from training data as input to make predictions
last_sequence = X_train[-1:]
forecasted_values = forecast(model, last_sequence, num_steps)
print(f"Forecasted values: {forecasted_values}")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('25JULY2024_1min.csv')  # for CSV file

# Ensure the timestamp column is in datetime format and set it as the index
df["TimeStamp"] = pd.to_datetime(df["TimeStamp"])
df.set_index("TimeStamp", inplace=True)

# Create a new index with 30-second intervals
new_index = pd.date_range(start=df.index.min(), end=df.index.max(), freq="10S")
df_10sec = df.reindex(new_index)

# Interpolate all columns (linear interpolation)
df_10sec = df_10sec.interpolate(method="linear")

# Reset the index and rename the timestamp column
df_10sec.reset_index(inplace=True)
df_10sec.rename(columns={"index": "TimeStamp"}, inplace=True)

# Save the result to a new CSV file
output_file = "25JULY2024_10S.csv"
df_10sec.to_csv(output_file, index=False)

print(f"Interpolated data saved to {output_file}")


print("\nConverted 10-second Interval Data:")
files.download('25JULY2024_10S.csv')
```

**Prepare data at required Sampling rate - Multiple files .xlsx or .csv**

```python
import pandas as pd
import os

# Define the directory containing the files
directory = '/content/drive/MyDrive/Segmented_1min'  # Update this to your folder path

# Define the resampling interval (e.g., '5T' for 5 minutes, '20T' for 20 minutes)
resample_interval = '5T'

# Loop through all the files in the directory
for filename in os.listdir(directory):
    # Check if the file is an Excel or CSV file
    if filename.endswith('.xlsx') or filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)

        # Load the data based on the file extension
        if filename.endswith('.xlsx'):
            df = pd.read_excel(file_path)
        elif filename.endswith('.csv'):
            df = pd.read_csv(file_path)

        # Ensure the timestamp column is in datetime format
        df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

        # Set the timestamp column as the index
        df.set_index('Time Stamp', inplace=True)

        # Resample the data to the specified interval and calculate the mean
        df_resampled = df.resample(resample_interval).mean()

        # Remove empty rows (rows with all NaN values)
        df_resampled = df_resampled.dropna(how='all')

        # Reset the index to include the timestamp again as a column
        df_resampled = df_resampled.reset_index()

        # Save the result to a new file with a modified name
        output_filename = f"{os.path.splitext(filename)[0]}_{resample_interval}_dropNA.csv"
        output_path = os.path.join(directory, output_filename)
        df_resampled.to_csv(output_path, index=False)

        print(f"Processed and saved: {output_filename}")
```

```python
def penalty_function(level, gas_flow):
    penalty = 0.0

    # Gas penalty
    if gas_flow > 75:
        penalty += (gas_flow - 75)**2
    elif gas_flow < 65:
        penalty += (65 - gas_flow)**2

    # Level bounds
    level_min = 35
    level_max = 55 if gas_flow < 75 else 35

    # Level penalty
    if level > level_max:
        penalty += (level - level_max)**2
    elif level < level_min:
        penalty += (level_min - level)**2

    return penalty

penalty_function(40, 7)
```

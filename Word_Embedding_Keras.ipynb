#**Embedding Layer**

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.preprocessing.text import one_hot
sentences= ["The glass of milk", 
            "the glass of juice",
            "the cup of tea",
            "I am a good boy",
            "I am a good developer",
            "understand the meaning of words",
            "your videos are good"]
#sentences
voc_size=10000 #tuning parameter
```

######**One Hot Representation**

```python
#onehot_repr = [one_hot(words,voc_size)for words in sentences]
onehot_repr=[]
for word in sentences:
  onehot_repr.append(one_hot(word,voc_size))

print(onehot_repr)
```

#####**Word Embedding Representation**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8 #tuning parameter
embedded_docs= pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
dim=15 #tuning parameter
model= Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
#model.compile('adam', 'mse')
model.summary()
```

```python
print(model.predict(embedded_docs))
```
